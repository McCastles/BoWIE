```
import time
import random
import numpy as np

def get_pretrained_glove( filename = 'glove.6B.50d.txt' ):

    foldername = '/content/drive/MyDrive/BOWIE/Data/GloVe/'
    # filename = 'glove.6B.50d.txt'

    embeddings = {}

    start = time.time()
    with open( foldername+filename ) as f:
        for line in f:
            s = line.index(' ')
            embeddings[ line[:s] ] = np.array( line[s+1:].split(' ') ).astype(np.float)

    print( f'Brought {len(embeddings.keys())} embeddings in {round(time.time()-start,1)}s' )
    randkey = random.choice( list(embeddings.keys()) )
    randval = embeddings[randkey]
    print( f'Random word embedding: {randkey} {randval}' )
    return embeddings, len(randval)
```
